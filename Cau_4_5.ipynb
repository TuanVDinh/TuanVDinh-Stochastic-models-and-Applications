```python
import numpy as np
from itertools import combinations


def enter_state_space():
	"""
        User entering state space.
    """
	
	# states
	states = []
	
	# number of states
	n = int(input("Enter number of states : "))
	
	# Enter the state space 
	for i in range(0, n):
		print(f"Enter state {i+1}:")
		state = input()
		states.append(state) 
	return states
	print("State space: ", states)

def enter_transition_matrix(num_states):
	"""
        User entering transition probability matrix 
    """
	
	transition_matrix = []
	
	print("Enter the transition probability matrix P:")

	for i in range(0, num_states):
		row = []
		for j in range(0, num_states):
			print(f"Enter P_{i}{j}:")
			row.append(float(input()))
		transition_matrix.append(row) 
	print("The transition probability matrix: ", transition_matrix)
	return transition_matrix

def states_need_check():
	"""
        User entering 2 states which need to be checked
    """
	print("Enter 2 states i and j to check:")
	print("Enter state i: ")
	i_state = input()
	print("Enter state j: ")
	f_state = input()
	return i_state, f_state
	
class MarkovChain(object):
    def __init__(self, transition_matrix, states):
        """
        Initialize the MarkovChain instance.

        Parameters
        ----------
        transition_matrix: 2-D array
            A 2-D array representing the probabilities of change of
            state in the Markov Chain.

        states: 1-D array
            An array representing the states of the Markov Chain. It
            needs to be in the same order as transition_matrix.
        """
        self.transition_matrix = np.atleast_2d(transition_matrix)
        self.states = states
        self.index_dict = {self.states[index]: index for index in
                           range(len(self.states))}
        self.state_dict = {index: self.states[index] for index in
                           range(len(self.states))}

    def is_accessible(self, i_state, f_state, check_up_to_depth=1000):
        """
        Check if state f_state is accessible from i_state.

        Parameters
        ----------
        i_state: str
            The state from which the accessibility needs to be checked.

        f_state: str
            The state to which accessibility needs to be checked.
        """
    
        counter = 0
        reachable_states = [self.index_dict[i_state]]
        for state in reachable_states:
            if counter == check_up_to_depth:
                return False
            if state == self.index_dict[f_state]:
                return True
            else:
                reachable_states.extend(np.nonzero(self.transition_matrix[state, :])[0])
            counter = counter + 1
            
    def do_communicate(self, i_state, f_state):
        """
        Check if state i_state and f_state communicate with each other.
		
		Parameters
		----------
		i_state: str
		
		f_state: str
		
		"""
        if self.is_accessible(i_state, f_state) and self.is_accessible(f_state, i_state):
            return True
        return False
        
    def is_irreducible(self):
        """
        Check if the Markov Chain is irreducible.
        """
        for (i, j) in combinations(self.states, 2):
            do_communicate = self.do_communicate(i, j)
            if  do_communicate == False:
                return False
        return True

if __name__ == "__main__": 
	
	states = enter_state_space()
	transition_matrix = enter_transition_matrix(len(states))
	markov_chain = MarkovChain(transition_matrix=transition_matrix,
	states=states)
	
	# =========== Cau 4 (a) ===========
	user_continue_access = True
	while user_continue_access:
		print("Check accessibility")
		i_state, f_state = states_need_check()
		is_accessible = markov_chain.is_accessible(i_state, f_state)
		if is_accessible == True:
			print(f"{f_state} is accessible from {i_state}")
		else:
			print(f"{f_state} is not accessible from {i_state}")
		print("Press Y to continue. Press N to stop checking.")
		continuing = input()
		continuing = continuing.lower()
		if continuing == "y":
			user_continue_access = True
		else:
			user_continue_access = False
	
	# =========== Cau 4 (b) ===========
	user_continue_com = True
	while user_continue_com:
		print("Check communication")
		i_state, f_state = states_need_check()
		do_communicate = markov_chain.do_communicate(i_state, f_state)
		if do_communicate == True:
			print(f"{i_state} and  {f_state} communicate with each other.")
		else:
			print(f"{i_state} and  {f_state} do not communicate with each other.")
		print("Press Y to continue. Press N to stop checking.")
		continuing = input()
		continuing = continuing.lower()
		if continuing == "y":
			user_continue_com = True
		else:
			user_continue_com = False
	
	# =========== Cau 5 ===========
	user_continue_irreducible = True
	while user_continue_irreducible:
		print("Check if a Markov chain is irreducible")
		states = enter_state_space()
		transition_matrix = enter_transition_matrix(len(states))
		markov_chain = MarkovChain(transition_matrix=transition_matrix,states=states)
		is_irreducible = markov_chain.is_irreducible()
		if is_irreducible == True:
			print("The Markov chain is irreducible")
		else:
			print("The Markov chain is reducible")
		
		print("Press Y to continue. Press N to stop checking.")
		continuing = input()
		continuing = continuing.lower()
		if continuing == "y":
			user_continue_irreducible = True
		else:
			user_continue_irreducible = False
	
```
